```python
import pandas as pd
```

```python
dam_weather_df = pd.read_csv('../Dam_weather_warning/weather_dam_data.csv',encoding='euc-kr',index_col=0)
```

```python
wrn_df = pd.read_csv('../Dam_weather_warning/HeavyRainStep2.csv',encoding='euc-kr',index_col=0)
```

```python
wrn_df.index = pd.to_datetime(wrn_df.index)
dam_weather_df.index = pd.to_datetime(dam_weather_df.index)
wrn_df.index.names = ['일시']
```

```python
wrn_df
```

```python
dam_weather_df.rename(columns={'합계 일사(MJ/m2)': '합계 일사량(MJ/m2)'},inplace=True)
```

```python
merged_df = pd.concat([dam_weather_df,wrn_df],axis=1)
```

```python
merged_df.fillna(0,inplace=True)
```
