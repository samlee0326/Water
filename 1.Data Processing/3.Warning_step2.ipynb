```python
import pandas as pd
```

```python
df = pd.read_csv('../Dam_weather_warning/HeavyRainWrn.csv',encoding='euc-kr')
```

```python
df
```

```python
df['Time_a'] = pd.to_datetime(df['Time_a'])
```

```python
df['Time_a_day'] =  df['Time_a'].values.astype('<M8[D]')
```

```python
df2 = df.iloc[:,2:]
```

```python
df2
```

```python
df2_col =df2.columns.tolist()
```

```python
df3 = df2.groupby('Time_a_day')[[c for c in df2_col[:-1]]].sum()
```

```python
#df3.to_csv('agg_result.csv',header=True,index=True)
```

```python
df3
```

```python
df3 = df3.asfreq(freq='1D')
```

```python
df3 = df3.fillna(value=0)
```

```python
df3
```

```python
df3.to_csv('../Dam_weather_warning/HeavyRainStep2.csv',header=True,encoding='euc-kr')
```
